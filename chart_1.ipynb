```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/ai dataset.csv')  # Replace 'your_file_path.csv' with your actual file path

# Calculate average pollutant levels
pollutants = ['PM10', 'PM2_5', 'NO2', 'SO2', 'O3']
average_pollutants = data[pollutants].mean()

# Bar graph: Average pollutant levels
plt.figure(figsize=(8, 5))
plt.bar(pollutants, average_pollutants, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.title('Average Levels of Air Pollutants')
plt.ylabel('Average Concentration (µg/m³)')
plt.xlabel('Pollutants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot: AQI vs Respiratory Cases
plt.figure(figsize=(8, 5))
plt.scatter(data['AQI'], data['RespiratoryCases'], alpha=0.5, color='teal')
plt.title('AQI vs Respiratory Cases')
plt.xlabel('Air Quality Index (AQI)')
plt.ylabel('Respiratory Cases')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
